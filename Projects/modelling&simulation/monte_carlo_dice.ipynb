```python

import random
import numpy as np
import matplotlib.pyplot as plt
```

### Monte Carlo simulation by rolling the dice. Suppose we roll two dice, and we want to predict the probability of getting the sum as 12.

```python
def roll():
    return random.randint(1,6)
```

```python
val = []
for i in range(0,10000000):
    sumRoll = roll() + roll()
    val.append(sumRoll)
```

```python
plt.hist(val,bins=11, density=True)
```
